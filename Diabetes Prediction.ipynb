```python
from sklearn.preprocessing import StandardScaler
sl = StandardScaler()
```

```python
X = df1.drop(['Outcome'],axis=1)
```

```python
y = df1['Outcome']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12)
```

```python
X_train = sl.fit_transform(X_train)
X_test = sl.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(fit_intercept=True, max_iter=1000)
clf.fit(X_train, y_train)
```

```python
clf.coef_
```

```python
clf.intercept_
```

```python
train_pred = clf.predict_proba(X_train)
test_pred = clf.predict_proba(X_test)
```

```python
train_class_preds = clf.predict(X_train)
test_class_preds = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
train_accuracy = accuracy_score(train_class_preds,y_train)
test_accuracy = accuracy_score(test_class_preds,y_test)

print("The accuracy on train data is ", train_accuracy)
print("The accuracy on test data is ", test_accuracy)
```

```python
labels = ['Non Diabetic', 'Diabetic']
cm = confusion_matrix(y_train, train_class_preds)
print(cm)

ax= plt.subplot()
sns.heatmap(cm, annot=True, ax = ax, fmt='.0f') #annot=True to annotate cells

# labels, title and ticks
ax.set_xlabel('Predicted labels')
ax.set_ylabel('True labels')
ax.set_title('Confusion Matrix')
ax.xaxis.set_ticklabels(labels)
ax.yaxis.set_ticklabels(labels)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
train_precision = precision_score(train_class_preds,y_train)
test_precision = precision_score(test_class_preds,y_test)

print(train_precision)
print(test_precision)
```

```python
train_recall = recall_score(train_class_preds,y_train)
test_recall = recall_score(test_class_preds,y_test)

print(train_recall)
print(test_recall)
```

```python
train_f1 = f1_score(train_class_preds,y_train)
test_f1 = f1_score(test_class_preds,y_test)

print(train_f1)
print(test_f1)
```

```python
from sklearn.model_selection import cross_validate
```

```python
logistic = LogisticRegression()
```

```python
scoring = ['accuracy','precision','recall']
scores = cross_validate(logistic,X, y, scoring = scoring, cv = 5, return_train_score=True,return_estimator=True,verbose = 10)
```

```python
scores['train_accuracy'].mean()
scores['test_accuracy'].mean()
scores['train_precision'].mean()
scores['test_precision'].mean()
scores['train_recall'].mean()
scores['test_recall'].mean()
```

```python
for model in scores['estimator']:
    print(model.coef_)
```

```python
from sklearn.model_selection import StratifiedKFold

# Create StratifiedKFold object.
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=12)
accu_train = []
accu_test = []

for train_index, test_index in skf.split(X, y):
	train_index = list(train_index)
	test_index = list(test_index)

	print(train_index)
	print(test_index)

	print(y.iloc[train_index].value_counts())
	print(y.iloc[test_index].value_counts())

	x_train_fold = X.iloc[train_index]
	x_test_fold = X.iloc[test_index]
	y_train_fold = y.iloc[train_index]
	y_test_fold = y.iloc[test_index]

	logistic.fit(x_train_fold, y_train_fold)

	train_class_preds = logistic.predict(x_train_fold)
	test_class_preds = logistic.predict(x_test_fold)

	accu_train.append(accuracy_score(train_class_preds,y_train_fold))
	accu_test.append(accuracy_score(test_class_preds,y_test_fold))
```

```python
print('List of train accuracy:', accu_train)
print('\nList of test accuracy:', accu_test)
print('\nMaximum train Accuracy That can be obtained from this model is:',
	max(accu_train)*100, '%')
print('\nMaximum test Accuracy That can be obtained from this model is:',
	max(accu_test)*100, '%')
print('\nMinimum train Accuracy:',
	min(accu_train)*100, '%')
print('\nMinimum test Accuracy:',
	min(accu_test)*100, '%')
print('\nOverall Train Accuracy:',
	(sum(accu_train)/len(accu_train))*100, '%')
print('\nOverall Test Accuracy:',
	(sum(accu_test)/len(accu_test))*100, '%')
```
